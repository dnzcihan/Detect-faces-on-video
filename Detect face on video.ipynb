```python
import cv2
import imageio
import imageio.plugins.ffmpeg
imageio.plugins.ffmpeg.download()
```

```python
face_classsifier = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
#eye_classsifier = cv2.CascadeClassifier('haarcascade-eye-default.xml')

def detect(frame):
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_classsifier.detectMultiScale(gray,1.3,5) 
    
    
    # dikdörtgen çizdirme
    
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        gray_faces = gray[y:y+h, x:x+w]
        color_faces = frame[y:y+h,x:x+w]
        #eyes = eye_classsifier.detectMultiScale(gray_faces,1.1,3) 

        #for(ex,ey,ew,eh),in eyes:
            #cv.rectangle(color_faces,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        
        
    return frame

reader = imageio.get_reader('1.mp4')
fps = reader.get_meta_data()['fps']
writter = imageio.get_writer('deneme.mp4',fps=fps)

for i,frame in enumerate(reader):
    frame = detect(frame)
    writter.append_data(frame)
    print(i)
writter.close()
```
